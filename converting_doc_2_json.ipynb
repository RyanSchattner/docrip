```python
import os
filepath='/content/drive/MyDrive/hrjson'
for subdir, dirs, files in os.walk("/content/drive/MyDrive/HRdata/HRProject"):
  for filename in files:
    if filename.endswith(".docx"): 
      print(os.path.abspath('/content/drive/MyDrive/HRdata/HRProject/{subdir}/{filename}'.format(filename=filename, subdir=subdir)))
      document = Document(os.path.abspath('{subdir}/{filename}'.format(filename=filename, subdir=subdir)))
      lines = [para.text for para in document.paragraphs]
      json_result = json.dumps(lines)
      with open('/content/drive/MyDrive/hrjson/{filename}.json'.format(filename=filename), 'w') as outfile:
        json.dump(json_result, outfile)
    else:
      print('fail')
      
```
